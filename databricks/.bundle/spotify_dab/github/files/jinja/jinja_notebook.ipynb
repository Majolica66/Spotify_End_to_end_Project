```python
parameters =[
    {
        "table":"spotify.silver.factstream",
        "alias":"fact",
        "cols":"fact.stream_id,fact.listen_duration"
    },
    {
        "table":"spotify.silver.dimuser",
        "alias":"dimuser",
        "cols":"dimuser.user_id,dimuser.user_name",
        "joins":"fact.user_id = dimuser.user_id"
    },
    {
        "table":"spotify.silver.dimtrack",
        "alias":"dimtrack",
        "cols":"dimtrack.track_id,dimtrack.track_name",
        "joins":"fact.track_id = dimtrack.track_id"
    }
    
]
```

```python
from jinja2 import Template
```

```python
query_text = """
            SELECT 
               {% for param in parameters %}
                    {{ param.cols }}
                        {% if not loop.last %}
                    ,
                        {% endif %}
                {% endfor %}
        FROM
        {% for param in parameters %}
            {% if loop.first %}
                {{ param['table'] }} AS {{ param['alias'] }}
            {% endif %}
        {% endfor %}
        {% for param in parameters %}
            {% if not loop.first %}
            LEFT JOIN
                {{ param['table'] }} AS {{ param['alias'] }}
            ON
                {{ param['joins'] }}
            {% endif %}
        {% endfor %}

"""
    
```

```python
jinja_sql_str = Template(query_text)
query = jinja_sql_str.render(parameters=parameters)  # Ensure 'parameters' is defined before this line.
print(query)
```

```python
display(spark.sql(query))
```
